```python
# Construindo gráficos
# Agrupando os dados:
groups = df_supervisionado.groupby('Classe')
```

```python
# Plotando o gráfico:
fig, ax = plt.subplots()
ax.margins(0.05)
for name, group in groups:
  ax.plot(group['Comp. Sepalas (cm)'], group['Larg. Sepalas (cm)'], marker='o', linestyle='', ms=12, label=name)

ax.legend()
plt.xlabel('Comp. Sepalas (cm)') # o que será apresentado no eixo x
plt.ylabel('Larg. Sepalas (cm)') # O que será apresentado no eixo y
plt.title('Comprimento e largura das Sepalas') # título do gráfico
leg = ax.legend(loc ='lower right') # Posicão da legenda
plt.show()
```

```python
# Plotando gráfico do dataset Não Supervisionado:
df_nao_supervisionado.plot.scatter(x='Comp. Sepalas (cm)', y='Larg. Sepalas (cm)',marker='o');
```

```python
# Criando um novo dataset a partir de um dataset existente.
df_supervisionado_mod=df_supervisionado.copy()
```

```python
# Criando uma nova coluna no novo dataset e adicionando dados calculados a partir do comprimento das sépalas:
df_supervisionado_mod['nova_coluna']=df_supervisionado['Comp. Sepalas (cm)']/100
```

```python
# Visualizando os primeiros dados do novo dataset:
df_supervisionado_mod.head()
```

```python
df_supervisionado.head()
```

```python
# Alterando o nome da coluna:
df_supervisionado_mod.rename(columns={'nova_coluna':'Comp. Sépalas (m)'})
```

```python
# Substituir o nome antigo da coluna para o novo nome definitivamente:
df_supervisionado_mod.rename(columns={'nova_coluna':'Comp. Sépalas (m)'},inplace=True)
```
